```
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```
X = np.linspace(-10,10,200)
y = 2.5 * X + 4 + np.random.normal(0,2,len(X))
```

```
px.scatter(X,y)
```

```
def plot_line(w,w0,X,y):
  sns.scatterplot(x=X,y=y)
  sns.lineplot(x=X,y=(w*X + w0),color='red')
  plt.show()

plot_line(2.5,4,X,y)
```

```
def loss_function(w,w0,X,y):
  # n = len(X)
  pridict = w*X + w0
  return np.sum((y-pridict)**2) + 100*w**2
```

```
def differnce(first1,last1,first2,last2,delta = 0.01):
  return (first1 - last1)/delta,(first2 - last2)/delta
```

```
def gradient(w,w0,X,Y):
    delta = 1e-5
    first1 = loss_function(w+delta,w0,X,Y)
    last1 = loss_function(w,w0,X,Y)
    first2 = loss_function(w,w0+delta,X,Y)
    last2 = loss_function(w,w0,X,Y)
    return differnce(first1,last1,first2,last2,delta)
```

```
loss_function(2.5,4,1,2)
```

```
def gradient_decent(w,w0,X,y,lr=0.0001,ITR=1000):
  for i in range(ITR):
    gradw,gradw0 = gradient(w,w0,X,y)
    w -= lr*gradw
    w0 -= lr*gradw0
    print(w,w0)
    if np.abs(gradw)<=0.01 and np.abs(gradw0) <= 0.01:
      plot_line(w,w0,X,y)
      print(f"Model Converged {i}")
      return w,w0
```

```
gradient_decent(100,200,X,y)
```
